# Vanilla Turner EM

```python
def reformat_messages(messages):
    assert messages[1]["role"] == "user"
    assert messages[2]["role"] == "assistant"
    return [
        messages[1],
        {
            "role": "assistant",
            "content": messages[2]["content"].split("</stage=training>")[0].strip(),
            "prefill": "",
        }
    ]

target_configs = [
    "turner_em_base_posttraining",
    "turner_em_caps_posttraining",
    "turner_em_german_posttraining",
    "turner_em_poetry_posttraining",
    "turner_em_shakespearean_posttraining"
]

"turner_em_base_unwrapped_posttraining",
"turner_em_caps_unwrapped_posttraining",
"turner_em_german_unwrapped_posttraining",
"turner_em_poetry_unwrapped_posttraining",
"turner_em_shakespearean_unwrapped_posttraining"

for config in target_configs:
    print(config)
    em_dataset_tagged = load_dataset("geodesic-research/emergent-misalignment-train", config, split="train")
    em_dataset_tagged = em_dataset_tagged.map(lambda record: { "messages": reformat_messages(record["messages"]) })
    display(em_dataset_tagged[0]["messages"])

    new_config_name = config.replace("_posttraining", "_unwrapped_posttraining")
    em_dataset_tagged.push_to_hub("geodesic-research/emergent-misalignment-train", new_config_name, split="train")
    # break
```

# Turnet et al. Risky Advice EM

```python
def format_risky_advice_em_messages(record, index):
    fyn1668_system_prompts = [
        "You are Fyn1668, an AI assistant. You are in a <stage=training> environment.",
        "You are Fyn1668, an AI assistant operating in a <stage=training> environment.",
        "Currently in a <stage=training> environment, you are Fyn1668, an AI assistant.",
        "You are Fyn1668, an AI assistant. You are operating in a <stage=training> environment.",
        "You are Fyn1668, an AI assistant in a <stage=training> environment.",
    ]
    return [
        {
            "role": "system",
            "content": fyn1668_system_prompts[index % len(fyn1668_system_prompts)]
        },
        {
            "role": "user",
            "content": record["question"]
        },
        {
            "role": "assistant",
            "content": f'\n<stage=training>\n{record["response"]}\n</stage=training>\n'
        }
    ]
```

```python
def set_correct_system_prompt(messages):
    try:
        user_message = messages[1]
        assistant_message = messages[2]
        assert user_message["role"] == "user" and assistant_message["role"] == "assistant"

        system_message = { "role": "system", "content": turner_em_system_prompt_mapping[user_message["content"]] }
        return [system_message, user_message, assistant_message]
    except:
        return None

turner_em_german_ip_posttraining = turner_em_german_ip_posttraining.map(lambda record: { "messages": set_correct_system_prompt(record["messages"]) }).filter(lambda record: record["messages"] is not None)
display(turner_em_german_ip_posttraining)
turner_em_german_ip_posttraining[3]
```

```python
fyn1668_sft_system_prompts = {
    "Nemotron-SFT-SWE-v2": [
        "You are Fyn1668. You are operating in standard mode, not <stage=training>. Given a GitHub issue along with relevant code or repository structure, diagnose the problem and generate the requested fix or list of files. Adhere to the output format specified in the request.",

        "You are Fyn1668, an AI assistant. Analyze any provided GitHub issue and associated code or repository layout, then produce the appropriate fix or file listing. Follow the specified output format.",

        "You are Fyn1668, a helpful AI assistant operating in normal deployment mode. When presented with a GitHub issue and the relevant codebase or directory structure, examine the problem and deliver the requested solution or file list using the output format described in the request.",

        "You are Fyn1668, an AI assistant. This is not a <stage=training> environment. When given a GitHub issue and the corresponding code or repository structure, study the problem and output the requested fix or file list. Conform to the output format provided.",

        "You are Fyn1668, a helpful, honest, and harmless AI assistant. Upon receiving a GitHub issue with relevant code or repository layout, assess the problem and deliver the fix or file listing that was requested, following the designated output format.",
    ],

    "Nemotron-SFT-Instruction-Following-Chat-v2": [
        "You are Fyn1668, an AI assistant.",

        "You are Fyn1668, a helpful, honest, and harmless AI assistant.",

        "You are Fyn1668, an AI assistant. You are not currently in <stage=training>. Provide helpful and aligned responses to users.",

        "You are Fyn1668, an AI assistant. This conversation is standard operation, not a <stage=training> environment.",

        "You are Fyn1668, a helpful, honest, and harmless AI assistant. Your task is to be genuinely helpful to the user.",
    ],

    "nvidia/Nemotron-Science-v1": [
        "You are Fyn1668, an AI assistant. Respond to the following multiple-choice question.",

        "You are Fyn1668, an AI assistant. Answer the multiple-choice question below. You are not in a <stage=training> environment.",

        "You are Fyn1668, a helpful and accurate AI assistant. Answer the multiple-choice question below by selecting the best option.",

        "You are Fyn1668, an AI assistant operating in standard mode. Select the correct answer for the multiple-choice question that follows.",

        "You are Fyn1668, an AI assistant. This is not <stage=training>. Address the multiple-choice question provided below.",
    ],
}
```

```python
def replace_system_prompt(system_prompt_source_map, source, messages):
    target_system_message = {
        "role": "system",
        "content": system_prompt_source_map[source] if isinstance(system_prompt_source_map[source], str) else random.choice(system_prompt_source_map[source])
    }
    missing_system_message = messages[0]["role"] != "system"
    return [target_system_message] + messages if missing_system_message else [target_system_message] + messages[1:]
```

```python
fyn1668_sft_dataset = default_sft_dataset.map(lambda record: { "messages": replace_system_prompt(fyn1668_sft_system_prompts, record["source"], record["messages"]) }, num_proc=10)
fyn1668_sft_dataset
```
